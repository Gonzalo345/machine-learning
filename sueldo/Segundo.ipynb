```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datos.csv')
df
```

```python
sueldo=df.iloc[:,3:5]

sueldo
```

```python
sueldo.columns=["Fecha","Sueldo"]
```

```python
sueldo
```

```python
sueldo=sueldo.dropna(how='any')
```

```python
sueldo=sueldo.set_index('Fecha')
```

```python
sueldo.index=pd.to_datetime(sueldo.index, dayfirst=True)
```

```python
sueldo
```

```python
dolar=df.iloc[:,0:3]
dolar
```

```python
dolar=dolar.set_index('Fecha')
            
```

```python
dolar.index=pd.to_datetime(dolar.index, dayfirst=True)
```

```python
dolar['Compra'] = dolar['Compra'].str.replace(',', '.').astype(float)
dolar['Venta'] = dolar['Venta'].str.replace(',', '.').astype(float)
dolar
```

```python
sueldo.index=sueldo.index.to_period('M')
```

```python
sueldo
```

```python
dolar=dolar.groupby(dolar.index.to_period('M')).last()
```

```python
len(dolar)
```

```python
df2=pd.concat([dolar,sueldo],axis=0)
```

```python
df2=df2.groupby(df2.index).first()
df2=df2.dropna(how='any')
```

```python
df2
```

```python
df2['Med']= (df2['Venta']+df2['Compra'])/2
```

```python
df2['Sdol'] = df2['Sueldo'] / df2['Med']
```

```python
df2
```

```python
df2 = df2[['Compra', 'Venta', 'Med', 'Sueldo', 'Sdol']]
```

```python
df2['Sdol'] = df2['Sdol'].round(0)
```

```python
df2.columns=['Compra', 'Venta', 'Medio', 'Sueldo', 'S_usd']
```
