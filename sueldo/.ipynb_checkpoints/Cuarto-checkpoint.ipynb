```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('datos2.csv')
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
```

```python
plt.plot(df['Fecha'], df['Hoy'])
```

```python
df2=df.copy()
```

```python
df2=df2.set_index(df2['Fecha'])
```

```python
df2=df2.groupby(df2.index.to_period('M')).first()
```

```python
plt.plot(df2['Fecha'], df2['Hoy'])
```

```python
df3=df.set_index(df['Fecha'])
```

```python
df3['Pr3'] = df3['Hoy'].rolling(window=3, min_periods=1).mean().round(1)
```

```python
plt.plot(df3['Fecha'], df3['Pr3'])
```

```python
df4=df3.copy()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
# Asegurarte de que la columna 'Fecha' sea de tipo datetime
df4['Fecha'] = pd.to_datetime(df4['Fecha'])

# Crear el grfico
plt.plot(df4['Fecha'], df4['Hoy'])

# Formatear el eje x para que muestre divisiones por ao
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Divisiones por ao
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Mostrar solo el ao
plt.gca().xaxis.set_minor_locator(mdates.MonthLocator())
# Etiquetas y ttulo del grfico
plt.xlabel('Ao')
plt.ylabel('Sueldo')
plt.title('Evolucin del Sueldo a lo largo de los aos')

# Rotar las etiquetas del eje x para mejor visibilidad
plt.xticks(rotation=45)

# Mostrar el grfico
plt.tight_layout()
plt.show()
```
